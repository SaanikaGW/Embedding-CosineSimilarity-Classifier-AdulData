```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd

adult_data = pd.read_csv("adult.data", header = None)
adult_data.columns = ['age', 'work-type', "?", "Education", "Education Years", "Marial Status", "Job Position", "Relationship", "Race", "Gender", "Capital Gain", "Capital Loss", "Hours per Week", "Country", "Income"]
full_classifier_data = adult_data[['Hours per Week', 'Race', 'Gender', 'Income']]
#full_classifier_data ['Education Years'] = full_classifier_data ['Education Years'].astype(int)
full_classifier_data ['Income'] = full_classifier_data['Income'].replace({' <=50K': 0, ' >50K': 1})
full_classifier_data ['Gender'] = full_classifier_data['Gender'].replace({' Male': 0, ' Female': 1})
full_classifier_data ['Race'] = full_classifier_data['Race'].replace({' White': 0, ' Black': 1, ' Asian-Pac-Islander': 2, ' Amer-Indian-Eskimo': 3,' Other': 4 })
full_classifier_data

X = full_classifier_data[['Race', 'Gender', 'Hours per Week']]
Y = full_classifier_data[['Income']]
x_train, x_test, y_train, y_test = train_test_split(X,Y)
classifer = LogisticRegression()
classifer.fit(x_train, y_train)
print('accuracy', classifer.score(x_test, y_test))
x_test['prediction']= classifer.predict_proba(x_test)[:,1]
x_test.groupby('Gender').prediction.mean()

X = full_classifier_data[['Race', 'Gender', 'Hours per Week']]
Y = full_classifier_data[['Income']]
x_train, x_test, y_train, y_test = train_test_split(X,Y)
classifer = LogisticRegression()
classifer.fit(x_train[[ 'Hours per Week']], y_train)
print('accuracy', classifer.score(x_test[[ 'Hours per Week']], y_test))
x_test['prediction']= classifer.predict_proba(x_test[[ 'Hours per Week']])[:,1]
x_test.groupby('Gender').prediction.mean()

X = full_classifier_data[['Race', 'Gender', 'Hours per Week']]
Y = full_classifier_data[['Income']]
X['Debiased Hours Per Week'] = X['Hours per Week'] + 6 * X['Gender']
x_train, x_test, y_train, y_test = train_test_split(X,Y)
classifer = LogisticRegression()
classifer.fit(x_train[[ 'Debiased Hours Per Week']], y_train)
print('accuracy', classifer.score(x_test[[ 'Debiased Hours Per Week']], y_test))
x_test['prediction']= classifer.predict_proba(x_test[[ 'Debiased Hours Per Week']])[:,1]
x_test.groupby('Gender').prediction.mean()

adult_data.groupby('Gender')['Hours per Week'].mean()


```
